```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

plt.plot(x, y1, 'b.', x, y2, 'k--')
plt.show()

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

```python
#Plot the noisy data (x, y1)
x = np.array([0,1,2,3,4,5])

#Plot the clean data (x, y2)
y = np.array([2.1, 2.9, 4.15, 4.98, 5.5, 6])

#Plot the line that fits the noisy data by the least square error
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

#Plotting
xp = np.linspace(-1, 6, 100)
plt.plot(x, y, '.', xp, p(xp))
plt.show()
```
