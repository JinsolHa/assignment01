# import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function: $f(x) = \cos(x) \cdot x$

```python
def myFunction(x):
    f = np.cos(x) * x
    return f
```

# define the derivative of my function: $f'(x) = -\sin(x) \cdot x + \cos(x)$

```python
def myDerivativeFunction(x):
    Df = - np.sin(x) * x + np.cos(x)
    return Df
```

# define the domain of the function: $x = [-15 : 0.1 : 15]$

```python
x = np.arange(-15, 15, 0.1)
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
```

# plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
